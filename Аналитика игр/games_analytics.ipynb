## Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры.  
Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

**Цель проекта** - выявить определяющие успешность игры закономерности.  
Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й.  
Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.  
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


**Описание данных:**
* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

В выборке присутствуют данные с 1980 по 2016 год.  
Данные за 2016 год могут быть неполными.

### Приведение типов


```python
df.info()
```

tbd в `user_score` значит "To Be Determined". Недостаточно оценок, либо игра вышла совсем недавно.  
С этим придется что-то сделать, так как далее потребуется анализ по оценкам пользователя, а с 'tbd' и пропусками не получится преобразовать тип в числовой.
Заменим 'tbd' в `user_score` нулями. Заполнять средним или любым другим вычисленным значением в данном случае мне кажется некорректным.
Пропуски в `user_score` и`critic_score`заполним по тому же принципу заполним нулями

*Наблюдение*: с помощью метода `astype` можно преобразовать из строки во float и при наличии пропусков. А при попытке преобразовать в int при наличии пропусков возникает ошибка и тип не меняется.

```python
# pd.to_numeric(df['user_score']) 
df.loc[df['user_score'] == 'tbd', ['user_score']] = 0
df['user_score'] = df['user_score'].fillna(0).astype('float32')
df['critic_score'] = df['critic_score'].fillna(0).astype('int32')

df.info()
```

Построим распределение по годам для платформ с наибольшими суммарными продажами и из раных поколений:  
PS2, XB,  PS3,  X360, PS, Wii  

```python
# сгруппируем данные по платформам, годам и продажам 
# построим графики с помощью цикла
# на графике отобразим суумарную выручку, но с помощью stacked выручки всех регионов
platform_to_plot = ['PS', 'Wii', 'PS2', 'XB', 'PS3', 'X360', 'PS4', 'XOne', 'PC']

for i in range(len(platform_to_plot)):
    (
        df[df['platform'] == platform_to_plot[i]]
        .groupby(['year_of_release'])['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'total_sales']
        .sum().reset_index()
        .plot(
            x = 'year_of_release', y = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],
            kind='bar', stacked=True, legend=True)
        .set_title(platform_to_plot[i])
    )
```

## Проверка гипотез
Задайте самостоятельно пороговое значение alpha.
Поясните:
* Как вы сформулировали нулевую и альтернативную гипотезы;
* Какой критерий применили для проверки гипотез и почему.


### Гипотеза 1

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Нулевая гипотеза H0: среднее значение пользовательского рейтинга игр для платформы Xbox One  равно среднему значению рейтинга для  PC

Альтернативная гипотеза H1: среднее значение пользовательского рейтинга игр для платформы Xbox One не равно среднему значению рейтинга для  PC

```python
# проведем Т-тест
# для теста возьмем выборки без пропусков в user_score, так как с ними Т-тест не работает

alpha = 0.05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

# result = st.ttest_ind(
#     df_actual.dropna(subset=['user_score'])[df_actual['platform'] == 'XOne']['user_score'],
#     df_actual.dropna(subset=['user_score'])[df_actual['platform'] == 'PC']['user_score'],
#     equal_var=True
# )
result = st.ttest_ind(
    df_actual[(df_actual['platform'] == 'XOne') & (df_actual['user_score'] != 0)]['user_score'],
    df_actual[(df_actual['platform'] == 'PC') & (df_actual['user_score'] != 0)]['user_score'],
    equal_var=False
)

print(result)
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
   
```

*Наблюдение*: T-тест не работает, если анализируемых данных есть пропуски

```python
#посмотрим на среднее выборок
print(
    'Cредний пользовательский рейтинг Xbox One:',
    df_actual[(df_actual['platform'] == 'XOne') & (df_actual['user_score'] != 0)]['user_score'].mean())
print(
    'Cредний пользовательский рейтинг PC:',
    df_actual[(df_actual['platform'] == 'PC') & (df_actual['user_score'] != 0)]['user_score'].mean())
```

### Гипотеза 2

Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза H0: среднее значение пользовательского рейтинга жанра Action  равно среднему значению рейтинга жанра Sports

Альтернативная гипотеза H1: среднее значение пользовательского рейтинга жанра Action не равно среднему значению рейтинга жанра Sports

Из выборок для проверки гипотез я убрал нули которыми заполнил пропуски (как и из корреляции), ведь по факту это не ноль в математическом смысле, когда критик или пользователь очень низко оценил игру, а ноль в смысле временной заглушки, которая имеет такое же значение, как и  'tbd' — пустоту или пропуск. В таких ситуациях, когда пропуск значит именно пропуск, с количественными столбцами зачастую лучшим решением будет ничего не делать. Потому что, как ты верно заметил, математические расчеты искажаются от заглушек. 
 

```python
# result = st.ttest_ind(
#     df_actual.dropna(subset=['user_score'])[df_actual['genre'] == 'Action']['user_score'],
#     df_actual.dropna(subset=['user_score'])[df_actual['genre'] == 'Sports']['user_score'],
#     equal_var=False
# )
result = st.ttest_ind(
    df_actual[(df_actual['genre'] == 'Action') & (df_actual['user_score'] != 0)]['user_score'],
    df_actual[(df_actual['genre'] == 'Sports') & (df_actual['user_score'] != 0)]['user_score'],
    equal_var=False
)

print(result)
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
print(
    'Средний пользовательский рейтинг жанра ACtion:',
    df_actual[(df_actual['genre'] == 'Action') & (df_actual['user_score'] != 0)]['user_score'].mean())
print('Дисперсия:', df_actual[(df_actual['genre'] == 'Action') & (df_actual['user_score'] != 0)]['user_score'].var())
print(
    'Средний пользовательский рейтинг жанра Sports:',
    df_actual[(df_actual['genre'] == 'Sports') & (df_actual['user_score'] != 0)]['user_score'].mean())
print('Дисперсия:', df_actual[(df_actual['genre'] == 'Sports') & (df_actual['user_score'] != 0)]['user_score'].var())
```
